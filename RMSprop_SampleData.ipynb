```python
import numpy as np
import pandas as pd
import math as mt
import random
```

```python
data=[[0.2,3.4],[0.4,3.8]]
data=pd.DataFrame(data,columns=['x','y'])
data
```

```python
Ns=data.shape[0]
epochs=500
eta=0.1
gama=0.9
eps=mt.pow(10,-8)
m=random.uniform(-1,1)
c=random.uniform(-1,1)
Em=0
Ec=0
```

```python
for iter in range(epochs):
  for sample in range(Ns):
    x=data["x"].iloc[sample]
    y=data["y"].iloc[sample]
    gm=-(y-m*x-c)*x
    gc=-(y-m*x-c)
    Em=gama*Em+(1-gama)*gm*gm
    Ec=gama*Ec+(1-gama)*gc*gc
    m=m-eta*gm/mt.sqrt(Em+eps)
    c=c-eta*gc/mt.sqrt(Ec+eps)
print(m)
print(c)
```

```python
yp=np.zeros(Ns,dtype=float)
for i in range(Ns):
  yp[i]=m*data["x"].iloc[i]+c
print(yp)
```

```python
ya=data["y"]
e=ya-yp
se=np.square(e)
mse=np.sum(se)/Ns
print(mse)
```
