# **Import all the libraries**

```python
import numpy as np
import pandas as pd
import math as mt
import random
```

# **Read data**

```python
data=[[0.2,3.4],[0.4,3.8]]
data=pd.DataFrame(data,columns=['x','y'])
data
```

# **Initialization**

```python
eta=0.9
epochs=500
m=random.uniform(-1,1)
c=random.uniform(-1,1)
Gm=0
Gc=0
eps=mt.pow(10,-8)
Ns=data.shape[0]
```

# **Adagrad Optimization algorithm implementation - batch size 1**

Adagrad Flowchart.png

```python
for iter in range(epochs):
  for sample in range(Ns):
    x=data["x"].iloc[sample]
    y=data["y"].iloc[sample]
    gm=-(y-m*x-c)*x
    gc=-(y-m*x-c)
    Gm=Gm+gm*gm
    Gc=Gc+gc*gc
    m=m-(eta/mt.sqrt(Gm+eps))*gm
    c=c-(eta/mt.sqrt(Gc+eps))*gc
print(m,c)
```

```python
yp=np.zeros(Ns,dtype=float)
for sample in range(Ns):
  yp[sample]=m*data["x"].iloc[sample]+c
print(yp)
```

```python
ya=data["y"]
e=ya-yp
se=np.square(e)
mse=np.sum(se)/Ns
print(e)
print(se)
print(mse)
```
