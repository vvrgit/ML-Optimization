**Import required libraries**

```python
import numpy as np
import math as mt
import random
import pandas as pd
```

***Import Data***

```python
data = pd.read_csv('/content/drive/MyDrive/AI Data/sampledata.csv')
data.head()
```

***Data Split***

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.X, data.Y, test_size = 0.20, random_state = 10)
```

***Initialization of ML model parameters***

```python
m=random.uniform(-10,10)
c=random.uniform(-10,10)
eta=0.9
max_iter=10000
v_m = 0
v_c = 0
gama = 0.9
```

***Gradient Calculation***

```python
def gradient(m,c,x,y):
  m_deriv=-1*(y-m*x-c)*x
  c_deriv=-1*(y-m*x-c)
  return m_deriv,c_deriv
```

***Training Using Stochastic Gradient Descent Optimizer***

```python
for iter in range(max_iter):
  for sample in range(x_train.shape[0]):
    del_m,del_c=gradient(m, c, x_train.iloc[sample], y_train.iloc[sample])
    v_m = gama * v_m - eta * del_m
    v_c = gama * v_c - eta * del_c
    m += v_m
    c += v_c

  print(f"Iteration: {iter}\tValue of m: {m}, \tValue of c: {c}")

print(f"\nThe local minima occurs at: {m}, {c}")
```

```python
q2=len(x_train)
error=0;
for j in range(q2):
    n2=m*x_train.iloc[j]+c
    error=error+(y_train.iloc[j]-n2)**2
error=error/q2
error
```

```python
q1=len(x_test)
error=0;
for j in range(q1):
    n2=m*x_test.iloc[j]+c
    error=error+(y_test.iloc[j]-n2)**2
error=error/q2
error
```
