```python
import numpy as np
import math as mt
import random
import pandas as pd
```

```python
data=pd.read_excel("/content/drive/MyDrive/AI Data/LR-Data.xlsx")
data.head()
```

```python
Fever_min=data['Fever'].min()
Fever_max=data['Fever'].max()
O2_min=data['O2'].min()
O2_max=data['O2'].max()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data)
data_norm=scaler.transform(data)
data_norm
```

```python
import pandas as pd
data_norm=pd.DataFrame(data_norm,columns=['Fever','O2','Cold','Cough','Covid'])
data_norm.head()
```

```python
from sklearn.model_selection import train_test_split
dataframe = data_norm.values
X, y = dataframe[:, :-1], dataframe[:, -1]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 10)
x_train
```

```python
m1=random.uniform(-10,10)
m2=random.uniform(-10,10)
m3=random.uniform(-10,10)
m4=random.uniform(-10,10)
c=random.uniform(-10,10)
eta=0.1
max_iter=10000
```

```python
for iter in range(max_iter):
  for sample in range(x_train.shape[0]):
    net=m1*x_train[sample,0]+m2*x_train[sample,1]+m3*x_train[sample,2]+m4*x_train[sample,3]+c
    yp=1/(1+np.exp(-net))
    m1=m1+eta*(y_train[sample]-yp)*x_train[sample,0]
    m2=m2+eta*(y_train[sample]-yp)*x_train[sample,1]
    m3=m3+eta*(y_train[sample]-yp)*x_train[sample,2]
    m4=m4+eta*(y_train[sample]-yp)*x_train[sample,3]
    c=c+eta*(y_train[sample]-yp)
print('model parameter m and c',m1,m2,m3,m4,c)
```

```python
Fever=input('enter Fever value')
O2=input('enter O2 value')
cold=input('enter cold value')
cough=input('enter cough value')
print('Fever=',Fever)
print('O2=',O2)
print('cold=',cold)
print('cough=',cough)


F_norm=(float(Fever)-Fever_min)/(Fever_max-Fever_min)
O2_norm=(float(O2)-O2_min)/(O2_max-O2_min)
net=m1*float(F_norm)+m2*float(O2_norm)+m3*float(cold)+m4*float(cough)+c
yp=1/(1+np.exp(-net))
print('Y=',round(yp))
if(yp>0.5):
  print("Covid Positive")
else:
  print("Covid Negative")
```
