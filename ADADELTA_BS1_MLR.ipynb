```python
import numpy as np
import pandas as pd
import math as mt
import random
```

```python
data=[[0.1,0.2,0.8],[0.2,0.1,0.7]]
df=pd.DataFrame(data,columns=['x1','x2','y'])
df
```

```python
Ns=df.shape[0]
gama=0.9
m1=random.uniform(-1,1)
m2=random.uniform(-1,1)
c=random.uniform(-1,1)
epochs=1000
Em1=0
Em2=0
Ec=0
Edm1=0
Edm2=0
Edc=0
dm1=0
dm2=0
dc=0
eps=mt.pow(10,-8)
```

```python
for iter in range(epochs):
  for sample in range(Ns):
    x1=df.iloc[sample,0]
    x2=df.iloc[sample,1]
    y=df.iloc[sample,2]

    gm1=-(y-m1*x1-m2*x2-c)*x1
    gm2=-(y-m1*x1-m2*x2-c)*x2
    gc=-(y-m1*x1-m2*x2-c)

    Em1=gama*Em1+(1-gama)*(gm1**2)
    Em2=gama*Em2+(1-gama)*(gm2**2)
    Ec=gama*Ec+(1-gama)*(gc**2)

    Edm1=gama*Edm1+(1-gama)*(dm1**2)
    Edm2=gama*Edm2+(1-gama)*(dm2**2)
    Edc=gama*Edc+(1-gama)*(dc**2)

    dm1=-(np.sqrt(Edm1+eps)/np.sqrt(Em1+eps))*gm1
    dm2=-(np.sqrt(Edm2+eps)/np.sqrt(Em2+eps))*gm2
    dc=-(np.sqrt(Edc+eps)/np.sqrt(Ec+eps))*gc

    m1=m1+dm1
    m2=m2+dm2
    c=c+dc
print(m1)
print(m2)
print(c)
```

```python
yp=np.zeros(Ns)
for i in range(Ns):
  yp[i]=m1*df.iloc[i,0]+m2*df.iloc[i,1]+c
print(yp)
ya=df["y"]
```

```python
e=ya-yp
se=np.square(e)
mse=np.mean(se)
print(mse)
```
