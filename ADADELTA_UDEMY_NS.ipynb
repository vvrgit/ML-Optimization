```python
import numpy as np
import pandas as pd
import random
import math as mt
```

```python
data=[[0.1,0.2,0.8],[0.2,0.1,0.7]]
data=pd.DataFrame(data,columns=['x1','x2','y'])
data
```

```python
Ns=data.shape[0]
gama=0.9
eps=mt.pow(10,-8)
m1=random.uniform(-1,1)
m2=random.uniform(-1,1)
c=random.uniform(-1,1)
epochs=2000
Em1=0
Em2=0
Ec=0
Edm1=0
Edm2=0
Edc=0
dm1=0
dm2=0
dc=0
```

```python
from IPython.core import macro
for iter in range(epochs):
  gm1=0
  gm2=0
  gc=0
  for sample in range(Ns):
    x1=data["x1"].iloc[sample]
    x2=data["x2"].iloc[sample]
    y=data["y"].iloc[sample]
    gm1=gm1+(y-m1*x1-m2*x2-c)*x1
    gm2=gm2+(y-m1*x1-m2*x2-c)*x2
    gc=gc+(y-m1*x1-m2*x2-c)

  # Gradient Calculation
  gm1=(-1/Ns)*gm1
  gm2=(-1/Ns)*gm2
  gc=(-1/Ns)*gc

  # Exponential Decaying Average of Squares of Gradients
  Em1=gama*Em1+(1-gama)*(gm1*gm1)
  Em2=gama*Em2+(1-gama)*(gm2*gm2)
  Ec=gama*Ec+(1-gama)*(gc*gc)

  # Exponential Decaying Average of Squares of step length
  Edm1=gama*Edm1+(1-gama)*(dm1*dm1)
  Edm2=gama*Edm2+(1-gama)*(dm2*dm2)
  Edc=gama*Edc+(1-gama)**(dc*dc)

  #Update step length
  dm1=-((mt.sqrt(Edm1+eps))/(mt.sqrt(Em1+eps)))*gm1
  dm2=-((mt.sqrt(Edm2+eps))/(mt.sqrt(Em2+eps)))*gm2
  dc=-((mt.sqrt(Edc+eps))/(mt.sqrt(Ec+eps)))*gc

  #Update model parameters
  m1=m1+dm1
  m2=m2+dm2
  c=c+dc


print(m1)
print(m2)
print(c)
```
