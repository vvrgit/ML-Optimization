# **Import Libraries**

```python
import numpy as np
import math as mt
import random
import pandas as pd
```

# **Load Data**

```python
data = pd.read_excel('/content/drive/MyDrive/AI Data/water.xlsx')
data.head()
```

# **Data Normalization**

```python
IP_min=data['X'].min()
IP_max=data['X'].max()
OP_min=data['Y'].min()
OP_max=data['Y'].max()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data)
data_norm=scaler.transform(data)
data_norm
```

```python
import pandas as pd
data_norm=pd.DataFrame(data_norm,columns=['IP','OP'])
data_norm.head()
```

# **Data Split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_norm.IP, data_norm.OP, test_size = 0.10, random_state = 10)
```

# **Model Initialization**

```python
m = random.uniform(-10,10) #Initial value of slope
c = random.uniform(-10,10) #Initial value of intercept
eta = 0.01 #Learning Rate
max_iter = 1000 #Maximum number of iterations
```

# **Gradeint Calculation**

```python
def gradient(m,c,x,y):
  m_deriv=-1*(y-m*x-c)*x
  c_deriv=-1*(y-m*x-c)
  return m_deriv,c_deriv
```

# **Model training using SGD**

```python
for iter in range(max_iter):
  for sample in range(x_train.shape[0]):
    del_m,del_c=gradient(m, c, x_train.iloc[sample], y_train.iloc[sample])
    m=m-eta*del_m
    c=c-eta*del_c
print('model parameter m and c',m,c)
```

# **Training Error**

```python
q2=len(x_train)
error=0;
for j in range(q2):
    n2=m*x_train.iloc[j]+c
    error=error+(y_train.iloc[j]-n2)**2
error=error/q2
error
```

# **Testing Error**

```python
q1=len(x_test)
error=0;
for j in range(q1):
    n2=m*x_test.iloc[j]+c
    error=error+(y_test.iloc[j]-n2)**2
error=error/q2
error
```

# **Model Deployment**

```python
X=input('enter value of pH')
X_norm=(float(X)-IP_min)/(IP_max-IP_min)
Y_norm=m*float(X_norm)+c
Y=Y_norm*(OP_max-OP_min)+OP_min
print('Bicarbonates=',Y)
```
