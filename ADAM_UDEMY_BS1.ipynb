```python
import numpy as np
import pandas as pd
import random
import math as mt
```

```python
data=[[0.1,1.1],[0.2,1.4]]
data=pd.DataFrame(data,columns=["x","y"])
Ns=data.shape[0]
data.head()
```

```python
epochs=50
eta=0.1
beta1=0.9
beta2=0.999
eps=mt.pow(10,-8)

m1=random.uniform(-1,1)
m2=random.uniform(-1,1)
c=random.uniform(-1,1)

Em11=0
Em21=0
Ec1=0

Em12=0
Em22=0
Ec2=0
t=0
```

```python
for iter in range(epochs):
  for sample in range(Ns):
    t=t+1
    #Gradeint Calculations
    x=data["x"].iloc[sample]
    y=data["y"].iloc[sample]
    gm1=-(y-m1*x*x-m2*x-c)*x*x
    gm2=-(y-m1*x*x-m2*x-c)*x
    gc=-(y-m1*x*x-m2*x-c)
    #Estimate First moment
    Em11=beta1*Em11+(1-beta1)*gm1
    Em21=beta1*Em21+(1-beta1)*gm2
    Ec1=beta1*Ec1+(1-beta1)*gc
    #Estimate second moment
    Em12=beta2*Em12+(1-beta2)*gm1*gm1
    Em22=beta2*Em22+(1-beta2)*gm2*gm2
    Ec2=beta2*Ec2+(1-beta2)*gc*gc
    #First moment correction
    Em11=Em11/(1-(beta1**t))
    Em21=Em21/(1-(beta1**t))
    Ec1=Ec1/(1-(beta1**t))
    #second moment correction
    Em12=Em12/(1-(beta2**t))
    Em22=Em22/(1-(beta2**t))
    Ec2=Ec2/(1-(beta2**t))
    #uPDATE model parameters
    m1=m1-(eta/(mt.sqrt(Em12)+eps))*Em11
    m2=m2-(eta/(mt.sqrt(Em22)+eps))*Em21
    c=c-(eta/(mt.sqrt(Ec2)+eps))*Ec1
print("m1=",m1)
print("m2=",m2)
print("c=",c)
```

```python
yp=np.zeros(Ns)
for i in range(Ns):
  yp[i]=m1*data["x"].iloc[i]*data["x"].iloc[i]+m2*data["x"].iloc[i]+c
print(yp)
```

```python
ya=data["y"]
e=ya-yp
se=np.square(e)
mse=np.mean(se)
print("MSE=",mse)
```
