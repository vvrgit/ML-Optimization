```python
import numpy as np
import math as mt
import pandas as pd
import random
```

```python
beta1=0.9
beta2=0.999
eta=0.1
eps=mt.pow(10,-8)
epochs=50

m1=random.uniform(-1,1);
m2=random.uniform(-1,1);
c=random.uniform(-1,1);

Em11=0
Em21=0
Ec1=0

Em12=0
Em22=0
Ec2=0
```

```python
data=[[0.1,0.11],[0.2,0.14]]
data=pd.DataFrame(data,columns=['x','y'])
Ns=data.shape[0]
data.head()
```

```python
for iter in range(epochs):
  for sample in range(Ns):
    # Read data corresponding to sample
    x=data.iloc[sample,0]
    y=data.iloc[sample,1]

    #Calculate Gradient
    gm1=-(y-m1*mt.pow(x,2)-m2*x-c)*mt.pow(x,2)
    gm2=-(y-m1*mt.pow(x,2)-m2*x-c)*x
    gc=-(y-m1*mt.pow(x,2)-m2*x-c)

    #compute first moment
    Em11=beta1*Em11+(1-beta1)*gm1
    Em21=beta1*Em21+(1-beta1)*gm2
    Ec1=beta1*Ec1+(1-beta1)*gc

    #compute second moment
    Em12=beta2*Em12+(1-beta2)*mt.pow(gm1,2)
    Em22=beta2*Em22+(1-beta2)*mt.pow(gm2,2)
    Ec2=beta2*Ec2+(1-beta2)*mt.pow(gc,2)

    #update first moment
    Em11=Em11/(1-beta1)
    Em21=Em21/(1-beta1)
    Ec1=Ec1/(1-beta1)

    #update second moment
    Em12=Em12/(1-beta2)
    Em22=Em22/(1-beta2)
    Ec2=Ec2/(1-beta2)

    #update parameters
    m1=m1-eta*Em11/(mt.sqrt(Em12)+eps)
    m2=m2-eta*Em21/(mt.sqrt(Em22)+eps)
    c=c-eta*Ec1/(mt.sqrt(Ec2)+eps)
print(m1)
print(m2)
print(c)
```

```python
yp=np.zeros(Ns)
for i in range(Ns):
  yp[i]=m1*mt.pow(data.iloc[i,0],2)+m2*data.iloc[i,0]+c
print(yp)
```

```python
ya=data["y"]
e=ya-yp
se=np.square(e)
mse=np.mean(se)
print(mse)
```
