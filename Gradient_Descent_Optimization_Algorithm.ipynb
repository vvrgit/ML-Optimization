# **Gradient Descent Optimization Algorithm**

image.png

```python
import numpy as np
x=np.random.uniform(-100,100)
eta=0.9
iter_max=200
```

```python
iter=1
while iter<=iter_max:
  slope=2*x
  step_length=-eta*slope
  x=x+step_length
  iter=iter+1
print("control variable=",round(x))
obj=x*x+5
print("objective function=",obj)
```

**Multi Varibale Optimization Using Gradient Descent Algorithm**

**Minimize x^2+y^2+10**

**Step1: Initialization**

```python
import numpy as np
import math as mt
eta=0.1
x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
itermax=100
iter=1
```

 ***Iterative Procedure***

```python
while iter<=itermax:
  delx=-eta*2*x
  dely=-eta*2*y
  x=x+delx
  y=y+dely
  iter=iter+1
```

**Print Solution**

```python
obj=mt.pow(x,2)+mt.pow(y,2)+10
print('Minimum value of objective function is',obj,'will get at x=',x,'y=',y)
```

```python
import numpy as np
import math as mt
eta=0.1
x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
itermax=100
iter=1
```

```python
while iter<=itermax:
  gradx=6*x
  grady=-5*mt.exp(-y)
  delx=-eta*gradx
  dely=-eta*grady
  x=x+delx
  y=y+dely
  iter=iter+1
print("x=",x)
print("y=",y)
```

```python
obj=3*mt.pow(x,2)+5*mt.exp(-y)+10
print('Minimum value of objective function is',obj,'will get at x=',x,'y=',y)
```
