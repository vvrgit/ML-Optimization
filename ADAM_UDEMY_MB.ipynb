```python
import numpy as np
import pandas as pd
import math as mt
import random
```

```python
data=[[0.1,1.1],[0.2,1.4],[0.3,1.9],[0.4,2.6]]
data=pd.DataFrame(data,columns=['x','y'])
data.head()
```

```python
eta=0.1
beta1=0.9
beta2=0.999
eps=mt.pow(10,-8)
epochs=50

m1=random.uniform(-1,1)
m2=random.uniform(-1,1)
c=random.uniform(-1,1)

t=0
Em11=0
Em21=0
Ec1=0

Em12=0
Em22=0
Ec2=0
```

```python
Ns=data.shape[0]
bs=2
Nb=int(Ns/bs)
batches=np.array_split(data,Nb)
batches
```

```python
for iter in range(epochs):
  for batch in range(Nb):
    t=t+1
    #gRADIENT cALUCLATION
    gm1=0;
    gm2=0;
    gc=0;
    for sample in range(bs):
      x=batches[batch].iloc[sample,0]
      y=batches[batch].iloc[sample,1]
      gm1=gm1+(y-m1*x*x-m2*x-c)*x*x
      gm2=gm2+(y-m1*x*x-m2*x-c)*x
      gc=gc+(y-m1*x*x-m2*x-c)
    gm1=(-1/bs)*gm1
    gm2=(-1/bs)*gm2
    gc=(-1/bs)*gc

    #Estimation of first moment
    Em11=beta1*Em11+(1-beta1)*gm1
    Em21=beta1*Em21+(1-beta1)*gm2
    Ec1=beta1*Ec1+(1-beta1)*gc

    #Estimation second moment
    Em12=beta2*Em12+(1-beta2)*gm1*gm1
    Em22=beta2*Em22+(1-beta2)*gm2*gm2
    Ec2=beta2*Ec2+(1-beta2)*gc*gc

    #fIRST MOMENT CORRECTion
    Em11=Em11/(1-mt.pow(beta1,t))
    Em21=Em21/(1-mt.pow(beta1,t))
    Ec1=Ec1/(1-mt.pow(beta1,t))

    #second moment correction
    Em12=Em12/(1-mt.pow(beta2,t))
    Em22=Em22/(1-mt.pow(beta2,t))
    Ec2=Ec2/(1-mt.pow(beta2,t))

    #update model parameters
    m1=m1-(eta/(mt.sqrt(Em12)+eps))*Em11
    m2=m2-(eta/(mt.sqrt(Em22)+eps))*Em21
    c=c-(eta/(mt.sqrt(Ec2)+eps))*Ec1

print(m1,m2,c)
```
